```python
sales_per_day = data.groupby('week_day')['quantity'].sum()
sales_per_day
sales_most_day = sales_per_day.idxmax()
print(f'Highest day sale is {sales_most_day}')
```

```python
size = data.groupby('size')['quantity'].sum()
size
```

```python
top_5_best_sell = data.groupby('pizza_type_id')['quantity'].sum().sort_values(ascending = False)

top_5_best_sell.head(5)
print(f"Top 5 best selling pizzas are: \n {top_5_best_sell.head(5)}")
```

```python
top_10_worst_sell = data.groupby('pizza_type_id')['quantity'].sum().sort_values(ascending = True)
top_10_worst_sell.head(10)
print(f'The Worst selling pizzas are : \n {top_10_worst_sell.head(10)}')
```

```python
revenue_pizza_type = data.groupby('pizza_type_id')['revenue'].sum().sort_values(ascending=False)

revenue_pizza_type
print(f'The revenue made by pizza type are: \n {revenue_pizza_type}')
```

```python
sales_day = data.groupby('week_day')['revenue'].sum()
print(f'The revenue made from sales per day \n {sales_day}')
```

```python
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday',
             'Friday', 'Saturday', 'Sunday']
sales_day = sales_day.reindex(day_order)
sales_day.plot(kind='barh', figsize = (8,4))
```

```python
monthly_revenue = data.groupby('month')['revenue'].sum().sort_values(ascending=False)
monthly_revenue

print(f"Monthly revenue is : \n {monthly_revenue}")
```

```python
monthly_sales = data.groupby('month')['quantity'].sum().sort_values(ascending = False)
monthly_sales
```

```python
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']
monthly_sales = monthly_sales.reindex(month_order)
monthly_sales.plot(kind='line', figsize = (14,6), marker= 'o')
plt.xticks(ticks=range(len(monthly_sales)), 
           labels=monthly_sales.index, 
           rotation=45)
plt.title('Monthly Sales Trend', fontweight = 'bold')
plt.xlabel('Month', fontweight = 'bold')
plt.ylabel('Quantity', fontweight = 'bold')
plt.tight_layout()
plt.show()
```
